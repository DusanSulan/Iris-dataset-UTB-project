# Projekt neuronové sítě pro IRIS dataset

```
 # import balíčků
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from keras.models import Sequential
from keras.layers.core import Dense
import keras.optimizers
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
import argparse
import pandas
from keras.utils import to_categorical
from keras import backend as K
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix

#sledované funkce počas treningu sítě v Kerasu
def sensitivity(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    return true_positives / (possible_positives + K.epsilon())

def specificity(y_true, y_pred):
    true_negatives = K.sum(K.round(K.clip((1-y_true) * (1-y_pred), 0, 1)))
    possible_negatives = K.sum(K.round(K.clip(1-y_true, 0, 1)))
    return true_negatives / (possible_negatives + K.epsilon())


dataset = pandas.read_csv("iris.csv", encoding='utf-8')
dataset = dataset.drop(['Id'], axis=1)
print(dataset.head())

import seaborn as sns
sns.set(style="ticks", color_codes=True)
sns.catplot(x="SepalWidthCm", y="SepalLengthCm", hue="Species",  data=dataset);

sns.set(style="ticks", color_codes=True)
sns.catplot(x="PetalWidthCm", y="PetalLengthCm", hue="Species",  data=dataset);
```

## Kód pro vytvoření, tréning modelů a vykreslení statistik:

```
# rozdelení na trénovací a validační část dat
(trainX, testX, trainY, testY) = train_test_split(dataset.drop(['Species'], axis=1),
dataset["Species"], test_size=0.25)

# labels to vectors
lb = LabelBinarizer()
trainY = lb.fit_transform(trainY)
testY = lb.transform(testY)

models = []
#model s RELU vrstvama, na konci nechávam softmax
model = Sequential()
model.add(Dense(32, input_shape=(4,), activation="relu"))
model.add(Dense(16, activation="relu"))
model.add(Dense(3, activation="softmax"))

models.append(model)
#model s tanh vrstvama, na konci nechávam softmax
model = Sequential()
model.add(Dense(32, input_shape=(4,), activation="tanh"))
model.add(Dense(16, activation="tanh"))
model.add(Dense(3, activation="softmax"))

models.append(model)
#model se sigmoid vrstvama, na konci nechávam softmax
model = Sequential()
model.add(Dense(32, input_shape=(4,), activation="sigmoid"))
model.add(Dense(16, activation="sigmoid"))
model.add(Dense(3, activation="softmax"))

models.append(model)

for model in models:
    # trening s RMS pro změnu kroku počas učení 
    optimizer = keras.optimizers.RMSprop(learning_rate=0.0005, rho=0.9) # SGD(0.0001)
    model.compile(loss="categorical_crossentropy", optimizer=optimizer,
                  metrics=["accuracy",sensitivity, specificity])
    num_epochs = 800 #počet tréninkových cyklů
    H = model.fit(trainX, trainY, validation_data=(testX, testY),
    epochs=num_epochs, batch_size=5) #trening

    #predikce validačných dat
    y_pred = model.predict_classes(testX) 

    #report konfuzní matice, F1, ...
    print(classification_report(testY.argmax(axis=1),
                y_pred,
                target_names=[str(x) for x in lb.classes_]))
    print(confusion_matrix(np.argmax(testY, axis=1), y_pred, labels=[0, 1, 2]))
    
    train_loss = H.history['loss']
    val_loss   = H.history['val_loss']
    
    sensitivity_y   = H.history['sensitivity']
    val_sensitivity_y =  H.history['val_sensitivity']
    
    specificity_y   = H.history['specificity']
    val_specificity_y = H.history['val_specificity']
    
    xc         = range(num_epochs)
    
    plt.figure()
    plt.plot(xc, train_loss, label='train_loss')
    plt.plot(xc, val_loss, label='val_loss')
    plt.plot(xc, sensitivity_y, label='sensitivity_y')
    plt.plot(xc, val_sensitivity_y, label='val_sensitivity')
    plt.plot(xc, specificity_y, label='specificity_y')
    plt.plot(xc, val_specificity_y, label='val_specificity_y')
    plt.legend()
```
